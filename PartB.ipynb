## Illustrating Failure Cases and Prompting Techniques with practical examples

### **1. Instruction-Based Prompt**  
**Failure Case:**  
The model misunderstands vague instructions.  

**Prompt Template:**  
> Please perform [specific task] using [parameters or guidelines].  

**Example:**  
Failure:  
> Summarize this document.  

Success:  
> Summarize this document in 100 words, focusing on the main arguments.

---

### **2. Open-Ended Question Prompt**  
**Failure Case:**  
The model gives irrelevant or overly generic responses.  

**Prompt Template:**  
> What are the main factors contributing to [topic]?  

**Example:**  
Failure:  
> What causes climate change?  
Response: "The environment."  

Success:  
> What are the main human and natural factors contributing to climate change?  
Response: "Human activities like fossil fuel burning and natural events like volcanic eruptions."

---

### **3. Step-by-Step Instruction Prompt**  
**Failure Case:**  
The model skips intermediate steps.  

**Prompt Template:**  
> Solve [problem] by breaking it down into steps.  

**Example:**  
Failure:  
> Solve 32 × 45.  
Response: "1440."  

Success:  
> Solve 32 × 45 by breaking it into smaller steps.  
Response: "32 × 45 = (30 + 2) × (40 + 5) = 1200 + 160 + 40 + 10 = 1440."

---

### **4. Fill-in-the-Blank Prompt**  
**Failure Case:**  
The model provides unrelated completions.  

**Prompt Template:**  
> Complete the sentence: [incomplete sentence].  

**Example:**  
Failure:  
> Complete the sentence: "The sky is..."  
Response: "delicious."  

Success:  
> Complete the sentence: "The sky is often described as..."  
Response: "blue on clear days."

---

### **5. Multiple-Choice Prompt**  
**Failure Case:**  
The model generates an answer outside the given options.  

**Prompt Template:**  
> Choose the best answer for [question]: [Option A], [Option B], [Option C].  

**Example:**  
Failure:  
> What is 2 + 2? (A) 3, (B) 4, (C) 5.  
Response: "6."  

Success:  
> Choose the best answer for 2 + 2: (A) 3, (B) 4, (C) 5.  
Response: "B) 4."

---

### **6. Example-Based Prompt**  
**Failure Case:**  
The model doesn't understand task formatting.  

**Prompt Template:**  
> Follow the format: [Example format].  

**Example:**  
Failure:  
> Translate "Hola."  
Response: "Hello, how are you?"  

Success:  
> Translate Spanish to English. Example: "Gracias" → "Thank you."  
> "Hola" → ?  
Response: "Hello."

---

### **7. Contrastive Prompt**  
**Failure Case:**  
The model fails to distinguish between options.  

**Prompt Template:**  
> Compare [Option A] and [Option B] based on [criteria].  

**Example:**  
Failure:  
> Compare cats and dogs.  
Response: "They are pets."  

Success:  
> Compare cats and dogs based on their maintenance and behavior.  
Response: "Cats require less maintenance, while dogs are more social."

---

### **8. Role-Based Prompt**  
**Failure Case:**  
The model doesn't adopt the assigned role.  

**Prompt Template:**  
> Act as [role] and perform [task].  

**Example:**  
Failure:  
> Act as a doctor and diagnose a fever.  
Response: "Fever means you're hot."  

Success:  
> Act as a doctor. Diagnose a patient with a fever of 101°F and sore throat.  
Response: "This could be a viral infection like the flu or strep throat. Seek medical advice."

---

### **9. Context-Setting Prompt**  
**Failure Case:**  
The model gives generic answers without understanding the context.  

**Prompt Template:**  
> Assume [specific context]. [Question/task].  

**Example:**  
Failure:  
> Why is water important?  
Response: "It's vital."  

Success:  
> Assume you are speaking to a group of farmers. Why is water important?  
Response: "Water is essential for irrigation and maintaining crop health."

---

### **10. Elaboration Prompt**  
**Failure Case:**  
The model provides overly brief answers.  

**Prompt Template:**  
> Explain [topic] in detail, including [specific aspects].  

**Example:**  
Failure:  
> Explain photosynthesis.  
Response: "Plants make food."  

Success:  
> Explain photosynthesis, including the role of sunlight, water, and chlorophyll.  
Response: "Photosynthesis is the process by which plants use sunlight, water, and carbon dioxide..."

---

### **11. Refinement Prompt**  
**Failure Case:**  
The model gives vague or incomplete responses.  

**Prompt Template:**  
> Improve [text/task] by [specific refinement].  

**Example:**  
Failure:  
> Rewrite this paragraph.  
Response: "No changes needed."  

Success:  
> Rewrite this paragraph to make it more concise.  

---

### **12. Hypothetical Scenario Prompt**  
**Failure Case:**  
The model doesn’t fully engage with the scenario.  

**Prompt Template:**  
> Imagine [scenario]. How would you [task]?  

**Example:**  
Failure:  
> Imagine you’re on Mars. Describe survival strategies.  
Response: "It's hard."  

Success:  
> Imagine you’re on Mars with limited resources. How would you survive the first week?  
Response: "Survival would require prioritizing oxygen production, water extraction, and shelter."

---

### **13. Completion-Based Prompt**  
**Failure Case:**  
The model generates irrelevant text.  

**Prompt Template:**  
> Finish this paragraph: [Partial text].  

**Example:**  
Failure:  
> Finish this paragraph: "Technology has advanced significantly..."  
Response: "The end."  

Success:  
> Finish this paragraph: "Technology has advanced significantly, particularly in..."  
Response: "fields like AI, robotics, and healthcare."

---

### **14. Correction Prompt**  
**Failure Case:**  
The model fails to recognize errors.  

**Prompt Template:**  
> Identify and correct errors in [text/task].  

**Example:**  
Failure:  
> Correct the sentence: "She go to school yesterday."  
Response: "Correct as is."  

Success:  
> Identify and correct grammatical errors: "She go to school yesterday."  
Response: "She went to school yesterday."

---

### **15. Summarization Prompt**  
**Failure Case:**  
The model provides overly general summaries.  

**Prompt Template:**  
> Summarize [text/task] in [specific format/length].  

**Example:**  
Failure:  
> Summarize the report.  
Response: "It's about sales."  

Success:  
> Summarize the sales report in 50 words, focusing on key trends.  

---

### **16. Debate Prompt**  
**Failure Case:**  
The model gives a one-sided argument.  

**Prompt Template:**  
> Debate [topic] by providing arguments for and against.  

**Example:**  
Failure:  
> Debate renewable energy.  
Response: "It's good."  

Success:  
> Debate renewable energy by listing its benefits and drawbacks.  

---

### **17. Role-Playing Prompt**  
**Failure Case:**  
The model breaks character.  

**Prompt Template:**  
> Role-play as [character] and respond to [situation].  

**Example:**  
Failure:  
> Role-play as a teacher explaining gravity.  
Response: "It's obvious."  

Success:  
> Role-play as a science teacher explaining gravity to a 10-year-old.  

---

### **18. Multi-Turn Interaction Prompt**  
**Failure Case:**  
The model doesn’t retain context between turns.  

**Prompt Template:**  
> Respond to [task] and maintain continuity in the dialogue.  

**Example:**  
Failure:  
> Q: What's 5 + 3?  
> A: 8.  
> Q: Multiply that by 2.  
> A: 6.  

Success:  
> Q: What's 5 + 3?  
> A: 8.  
> Q: Multiply that by 2.  
> A: 16.  

---

### **19. Diagnostic Prompt**  
**Failure Case:**  
The model fails to identify issues in [input].  

**Prompt Template:**  
> Analyze [input/task] and diagnose any issues.  

**Example:**  
Failure:  
> Analyze the program's output.  
Response: "It's fine."  

Success:  
> Analyze this Python code for errors: `print(2 + )`.  
Response: "There's a syntax error due to the incomplete addition."

---

### **20. Brainstorming Prompt**  
**Failure Case:**  
The model provides limited ideas.  

**Prompt Template:**  
> Generate [number] ideas for [task/topic].  

**Example:**  
Failure:  
> Generate ideas for starting a business.  
Response: "Open a shop."  

Success:  
> Generate 5 innovative business ideas in the tech industry.  

---

### **21. Specific Format Prompt**  
**Failure Case:**  
The model doesn’t follow the format.  

**Prompt Template:**  
> Provide the response in [specific format].  

**Example:**  
Failure:  
> List reasons for exercise.  
Response: "Exercise is good for health."  

Success:  
> List 3 reasons for exercise:  
> 1. Improves cardiovascular health.  
> 2. Enhances mental well-being.  
> 3. Increases energy levels.
---
